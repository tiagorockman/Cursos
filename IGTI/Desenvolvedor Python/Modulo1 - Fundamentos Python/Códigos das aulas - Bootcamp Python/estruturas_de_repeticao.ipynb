```
#contando a quantidade de elementos inteiros existem em uma lista
string_inteiros=input("Entre com uma lista de inteiros.  ")
lista_inteiros=string_inteiros.split()
contador= 0 #aqui definimos o nosso acumulador e inicializamos a contagem
for inteiro in lista_inteiros:
  contador=contador+1
print("Existem %.d inteiros na lista" % contador)
```

```
#calculando o fatorial (!) de um número 
n=int(input("Entre com um número não negativo: "))
fatorial=1
for i in range(1,n+1):
  fatorial=fatorial*i
print(str(n)+"! = ",fatorial)
```

**Lendo linha a linha de um arquivo com o While**

image.png

```
#contando a quantidade de linhas em um arquivo com o while
nome_do_arquivo='carlos_drummond.txt'
poema= open(nome_do_arquivo)
numero_de_linhas= 0
#lê a primeira linha do arquivo
linha=poema.readline().rstrip()
while linha !="FIM":
  numero_de_linhas=numero_de_linhas+1
  linha=poema.readline().rstrip()
print(nome_do_arquivo+" possui "+str(numero_de_linhas)+" linhas.")
```

```
#lendo as linhas do arquivo com o while
nome_do_arquivo='carlos_drummond.txt'
poema= open(nome_do_arquivo)
lista_linhas= []
#lê a primeira linha do arquivo
linha=poema.readline().rstrip()
while linha !="FIM":
  lista_linhas.append(linha)
  linha=poema.readline().rstrip()
for i in lista_linhas:
  print(i)
```

```
#while com diferentes condições
i = 1
while i <= 6:
  print(i)
  i += 1  #acumulador
```

**Break em um loop while**

```
#saindo de um loop mesmo quando a condição é verdadeira
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

**Continue em um loop while**

```
#utilizando o continue - para e continua aqui
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```
