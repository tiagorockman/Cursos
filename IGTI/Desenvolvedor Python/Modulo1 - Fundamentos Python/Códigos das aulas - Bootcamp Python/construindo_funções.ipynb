#**Pontos a serem observados quando construímos funções**

1.   **Qual será o nome da minha função?** O nome deve fazer sentido para o código que está desenvolvendo. O nome deve mostrar o que a função executa.
2.   **Quais argumentos a função deve receber?** Devemos definir quais os valores devem ser passados para a função.
3.   **O que a função deve fazer?** Devemos definir o papel da função em nosso programa, ou seja, o que ela deve fazer com os valores passados.
4.   **O que a função deve retornar?** Após realizar o processamento, qual o valor esperamos receber da função?



```
#escreva uma função que retorne o valor absoluto de um número 

# 1 - o nome da função será valor_absoluto
# 2 - a função vai receber um número (inteiro ou float)
# 3 - a função deve calcular o valor absoluto de do número passado
# 4 - a função deve retornar o valor absoluto
def valor_absoluto(num):
    """Esta função retorna o valor
    absoluto de um número"""

    if num >= 0:
        return num
    else:
        return -num


print(valor_absoluto(20))

print(valor_absoluto(-4.5))
```

# **Escopo de um código Python**

```
#seja esta função
def minha_funcao():  #não recebe argumento
  variavel_interna = 300             #variável definida no escopo da função
  print(variavel_interna)

minha_funcao()
```

```
#tentando acessar a variável fora da função
print(variavel_interna)
```

```
#acessando a variável dentro dentro de outra função
def minha_funcao():
  variavel_interna = 300  #variável interna da função
  def minha_funcao_interna():  #função interna da função "minha_funcao"
    print(variavel_interna)
  minha_funcao_interna() #chamando a função interna dentro da função

minha_funcao()
```

**Escopo Global**

```
variavel_externa = 300

def minha_funcao():
  print(variavel_externa)

minha_funcao()

print(variavel_externa)
```

```
#modificando variáveis globais
variavel = 300

def minha_funcao():
  variavel = 200
  print(variavel)

minha_funcao()

print(variavel)
```

```
#utilizando a palavra global
def minha_funcao():
  global variavel_global  #utiliza a palavra-chave global para definir uma variável global no escopo da função
  variavel_global = 300

minha_funcao()

print(variavel_global)
```

```
#trocando o valor da variável global dentro da função altera o valor da variável
x = 300

def minha_funcao():
  global x
  x = 200

minha_funcao()

print(x)
```

```
#função que inverte o valor da string passada
def inverte_string(valor_string):
  resultado=" "
  for caracter in valor_string:
    resultado = caracter +resultado

  return resultado
```

```
#utilizando a função definida

string=input("Entre com alguma string: ")
while string.strip() !="":
  print("O inverso de ",string, "é ", inverte_string(string))
  string=input("Entre com outra string ou pressione ENTER para sair: ")
```

```
#construindo o programa via bottom-up
def verifica_divisibilidade(x,y):
  return int(y) % int(x)==0
```

```
#em um modelo de bottom-up
#definindo uma função que retorna a lista com os valores da lista que são divisores de x
def elementos_divisiveis(x,lista):
  resultado=[]

  for elemento in lista:
    if verifica_divisibilidade(elemento,x):
      resultado.append(elemento)
  
  return resultado
```

```
#função que recebe uma lista de valores e retorna quem são os divisores
def divisores_em_lista(lista):
  for elemento in lista:
    print(elemento, " é divisível por ", end=" ")
    elementos=elementos_divisiveis(elemento,lista)
    for f in elementos:
      print(f, end=" ")
    print()
```

```
s= input("Entre com uma sequência de valores separados por espaço: ")
lista_valores=[]
for x in s.split():
  lista_valores.append(int(x))
divisores_em_lista(lista_valores)
```

# **Funções recursivas**

image.png



*   Função simplifica o problema em casos mais simples;
*   A própria função resolve os casos menores;
*   Resolve o caso mais simples (caso base);
*   Retorna a solução para o caso base;
*   Resolve o problema mais complexo.



image.png

```
#função fatorial - (n! = n *(n-1) * (n-2) * ....* 1)  - !1=1  !0=1
# 5! = 5*4!
# 4! = 4*3!
# 3! = 3*2!
# 2! = 2*1!
# 1! = 1
def fatorial(n):
  if n==0:
    return 1

  return n*fatorial(n -1)

print(fatorial(10))
```

```
#utilizando a função main para deixar o código "mais limpo"  - fonte: realpython.com
from time import sleep


def process_data(data):
    print("Iniciando o processamento de algum dado...")
    modified_data = data + " o dado foi modificado"
    sleep(3)
    print("Processamento do dado terminado.")
    return modified_data

def read_data_from_web():
    print("Lendo dados da web")
    data = "Dados da web lidos"
    return data

def write_data_to_database(data):
    print("Armazenando dados em um database")
    print(data)

def main():
    data = read_data_from_web()
    modified_data = process_data(data)
    write_data_to_database(modified_data)

if __name__ == "__main__":
    main()
```
